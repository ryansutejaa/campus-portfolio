```python
data_ump_performance = df_performance_uni_mean_transposed['UMP']
data_uncen_performance = df_performance_uni_mean_transposed['UNCEN']
data_undana_performance = df_performance_uni_mean_transposed['UNDANA']
data_unhena_performance = df_performance_uni_mean_transposed['UNHENA']
data_unimor_performance = df_performance_uni_mean_transposed['UNIMOR']
data_unimuda_performance = df_performance_uni_mean_transposed['UNIMUDA']
data_untrib_performance = df_performance_uni_mean_transposed['UNTRIB']
data_unvic_performance = df_performance_uni_mean_transposed['UNVIC']
```

### Concat importance and performance grouped by university

```python
df_ump = pd.concat([data_ump_performance, data_ump_importance], axis=1, keys=['performance', 'importance'])
df_uncen = pd.concat([data_uncen_performance, data_uncen_importance], axis=1, keys=['performance', 'importance'])
df_undana = pd.concat([data_undana_performance, data_undana_importance], axis=1, keys=['performance', 'importance'])
df_unhena = pd.concat([data_unhena_performance, data_unhena_importance], axis=1, keys=['performance', 'importance'])
df_unimor = pd.concat([data_unimor_performance, data_unimor_importance], axis=1, keys=['performance', 'importance'])
df_unimuda = pd.concat([data_unimuda_performance, data_unimuda_importance], axis=1, keys=['performance', 'importance'])
df_untriv = pd.concat([data_untrib_performance, data_untrib_importance], axis=1, keys=['performance', 'importance'])
df_unvic = pd.concat([data_unvic_performance, data_unvic_importance], axis=1, keys=['performance', 'importance'])
```

```python
df_list = [df_ump, df_uncen, df_undana, df_unhena, df_unimor, df_unimuda, df_untriv, df_unvic]
```

### Calculating CSI

```python
for df_univ in df_list:
  #CSI
  df_univ['WF'] = df_univ['importance']/sum(df_univ['importance'])
  df_univ['WS'] = df_univ['performance']*df_univ['WF']
```

```python
univ = ['UMP','UNCEN','UNDANA','UNHENA','UNIMOR','UNIMUDA','UNTRIV','UNVIC']
```

```python
csi_scores = []
for df_univ in df_list:
    csi = (sum(df_univ['WS']) / 4)
    csi = round(csi * 100, 3)
    csi_scores.append(csi)

data_csi_univ = {'University': univ, 'CSI': csi_scores}
df_csi_univ = pd.DataFrame(data_csi_univ)

print(df_csi_univ)
```

## Hypothesis testing (t-test)

```python
from scipy import stats
import numpy as np

# Data CSI universitas
csi_data = df_csi_univ.CSI

# Target CSI Anda
target_csi = 80

# Hitung rata-rata nilai CSI
mean_csi = np.mean(csi_data)

# Lakukan uji hipotesis
t_statistic, p_value = stats.ttest_1samp(csi_data, target_csi, alternative='greater')

# Tampilkan hasil
print("Mean CSI:", mean_csi)
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
```

If pvalue <0.05, h0 rejected

H0 = mean data <= target csi

H1 = mean data > target csi
